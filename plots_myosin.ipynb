```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# 1. Your Presentation Data
data = {
    "Reaction_Time": [2.5, 5.0, 7.5, 2.5, 5.0, 7.5, 2.5, 5.0, 7.5],
    "Interaction_Value": [0.40, 0.50, 0.70, 0.30, 0.60, 0.50, 0.50, 0.40, 0.60],
    "Ligand": ["ADP", "ADP", "ADP", "Ca2+", "Ca2+", "Ca2+", "Mg2+", "Mg2+", "Mg2+"]
}

df = pd.DataFrame(data)

# 2. Styling (Matching the 'paletteer' Frida Kahlo vibe)
plt.figure(figsize=(9, 6))
sns.set_theme(style="whitegrid")
custom_palette = {"ADP": "#E6956F", "Ca2+": "#004D40", "Mg2+": "#7C9865"}

# 3. Plotting
sns.lineplot(
    data=df, x="Reaction_Time", y="Interaction_Value", 
    hue="Ligand", marker="o", markersize=10, linewidth=2.5, 
    palette=custom_palette
)

# 4. Labels and Title
plt.title("Myosin Interaction Dynamics vs Reaction Time", fontsize=16, fontweight="bold", pad=15)
plt.suptitle("Analysis of Argopecten irradians motor protein binding metrics", color="gray", y=0.92)
plt.xlabel("Reaction Time (Seconds)", fontsize=12, labelpad=10)
plt.ylabel("Interaction Value", fontsize=12, labelpad=10)
plt.legend(title="Ligand", title_fontsize='11', fontsize='10')
plt.xticks([2.5, 5.0, 7.5])

# 5. Save and Show
plt.savefig("ligand_potency_plot.png", dpi=300, bbox_inches="tight")
print("[✓] Plot saved successfully as 'ligand_potency_plot.png'")
plt.show()
```
